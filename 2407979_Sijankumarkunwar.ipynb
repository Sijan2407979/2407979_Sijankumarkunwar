```python
#Task - 3 - Outlier Detection:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file for South Asian countries
df = pd.read_csv('South_Asian_Countries_Happiness.csv')

# Function to identify outliers using the 1.5 * IQR rule
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Step 1: Identify outliers based on Score and GDP per Capita
outliers_score = identify_outliers(df, 'score')
outliers_gdp = identify_outliers(df, 'Log GDP per capita')

# Combine both outliers for easy identification
outliers_combined = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

# Step 2: Create a scatter plot with GDP per Capita vs. Score
plt.figure(figsize=(10, 6))

# Plot all countries
sns.scatterplot(x='Log GDP per capita', y='score', data=df, label='Normal', color='blue')

# Highlight outliers in a different color
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers_combined, label='Outliers', color='red')

# Add titles and labels
plt.title('Scatter Plot: GDP per Capita vs Score (Outliers Highlighted)', fontsize=16)
plt.xlabel('GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()

# Step 3: Print out the outlier countries
print("Outlier countries based on Score and GDP per Capita:")
print(outliers_combined[['Country name', 'Log GDP per capita', 'score']])
```

```python
#Task 4 Happiness Disparity:
import pandas as pd

# Load the South Asia and Middle East datasets
south_asia_df = pd.read_csv('South_Asian_Countries_Happiness.csv')
middle_east_df = pd.read_csv('Middle_Eastern_Countries_Happiness.csv')

# Function to calculate range and coefficient of variation
def compute_variability(df, region_name):
    max_score = df['score'].max()
    min_score = df['score'].min()
    range_score = max_score - min_score

    mean_score = df['score'].mean()
    std_score = df['score'].std()
    cv_score = (std_score / mean_score) * 100

    return {
        'Region': region_name,
        'Max Score': max_score,
        'Min Score': min_score,
        'Range': range_score,
        'Mean': mean_score,
        'Standard Deviation': std_score,
        'Coefficient of Variation': cv_score
    }

# Compute statistics for both regions
south_asia_stats = compute_variability(south_asia_df, "South Asia")
middle_east_stats = compute_variability(middle_east_df, "Middle East")

# Combine results into a DataFrame
results_df = pd.DataFrame([south_asia_stats, middle_east_stats])

# Display results
print("Happiness Disparity Between South Asia and Middle East:")
print(results_df)

# Determine which region has greater variability
greater_variability = (
    "South Asia" if south_asia_stats['Coefficient of Variation'] > middle_east_stats['Coefficient of Variation']
    else "Middle East"
)

print(f"\nThe region with greater variability in Happiness Scores is: {greater_variability}")
```

```python
#Task 6 Outlier Detection:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
south_asia_df = pd.read_csv('South_Asian_Countries_Happiness.csv')
middle_east_df = pd.read_csv('Middle_Eastern_Countries_Happiness.csv')

# Function to detect outliers using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers for Score and GDP per Capita in both regions
score_outliers_sa = detect_outliers(south_asia_df, 'score')
gdp_outliers_sa = detect_outliers(south_asia_df, 'Log GDP per capita')

score_outliers_me = detect_outliers(middle_east_df, 'score')
gdp_outliers_me = detect_outliers(middle_east_df, 'Log GDP per capita')

# Combine outliers for each region
outliers_sa = pd.concat([score_outliers_sa, gdp_outliers_sa]).drop_duplicates()
outliers_me = pd.concat([score_outliers_me, gdp_outliers_me]).drop_duplicates()

# Print identified outliers
print("Outlier Countries in South Asia based on Score and GDP per Capita:")
print(outliers_sa[['Country name', 'score', 'Log GDP per capita']])

print("\nOutlier Countries in Middle East based on Score and GDP per Capita:")
print(outliers_me[['Country name', 'score', 'Log GDP per capita']])

# Scatter plot to visualize outliers
def plot_outliers(df, outliers, region_name):
    plt.figure(figsize=(8, 6))
    # Scatter plot for all countries
    sns.scatterplot(x='Log GDP per capita', y='score', data=df, color='blue', label='Countries')
    # Highlight outliers
    sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', s=100, label='Outliers')

    plt.title(f"Outliers in {region_name} (GDP per Capita vs Score)")
    plt.xlabel("GDP per Capita")
    plt.ylabel("Score")
    plt.legend()
    plt.show()

# Plot for South Asia
plot_outliers(south_asia_df, outliers_sa, "South Asia")

# Plot for Middle East
plot_outliers(middle_east_df, outliers_me, "Middle East")
```
