```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression as SklearnLR, Ridge, Lasso
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.impute import KNNImputer

# Load the dataset
file_path = '/content/drive/My Drive/Aiportfolio/global-data-on-sustainable-energy .csv'
df = pd.read_csv(file_path)

# Step 1: Handle missing values using KNN Imputer
imputer = KNNImputer(n_neighbors=5)
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)

# Ensure target column is also handled
df_imputed['Access to electricity (% of population)'] = df['Access to electricity (% of population)'].fillna(df['Access to electricity (% of population)'].median())

# Step 2: Feature Engineering (add interaction terms and polynomial features)
X = df_imputed.drop(columns=['Access to electricity (% of population)'])
y = df_imputed['Access to electricity (% of population)']

# Polynomial Features to capture non-linear relationships
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 3: Feature Scaling
scaler = StandardScaler()
X_poly_scaled = scaler.fit_transform(X_poly)

# Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_poly_scaled, y, test_size=0.2, random_state=42)

# Step 5: Apply Regularization (Ridge or Lasso Regression)
ridge = Ridge(alpha=1)
lasso = Lasso(alpha=0.01)

# Train the models
ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)

# Step 6: Evaluate the models using R-squared and MSE
ridge_r2 = ridge.score(X_test, y_test)
lasso_r2 = lasso.score(X_test, y_test)
ridge_mse = mean_squared_error(y_test, ridge.predict(X_test))
lasso_mse = mean_squared_error(y_test, lasso.predict(X_test))

# Output the R-squared and MSE for both models
print(f'R-squared (Ridge): {ridge_r2}')
print(f'Mean Squared Error (Ridge): {ridge_mse}')
print(f'R-squared (Lasso): {lasso_r2}')
print(f'Mean Squared Error (Lasso): {lasso_mse}')

# Step 7: Hyperparameter Optimization (using GridSearchCV for Ridge and Lasso)
param_grid = {
    'alpha': [0.01, 0.1, 1, 10, 100]
}

ridge_grid = GridSearchCV(Ridge(), param_grid, cv=5)
lasso_grid = GridSearchCV(Lasso(), param_grid, cv=5)

ridge_grid.fit(X_train, y_train)
lasso_grid.fit(X_train, y_train)

print(f"Best Ridge alpha: {ridge_grid.best_params_}")
print(f"Best Lasso alpha: {lasso_grid.best_params_}")

# Step 8: Cross-validation and Model Comparison
ridge_cv_score = cross_val_score(ridge_grid.best_estimator_, X_train, y_train, cv=5, scoring='r2')
lasso_cv_score = cross_val_score(lasso_grid.best_estimator_, X_train, y_train, cv=5, scoring='r2')

print(f'Cross-validated R2 score (Ridge): {np.mean(ridge_cv_score)}')
print(f'Cross-validated R2 score (Lasso): {np.mean(lasso_cv_score)}')

# Step 9: Plot results
plt.figure(figsize=(8, 6))
plt.scatter(y_test, ridge.predict(X_test), label='Ridge Predictions', alpha=0.6)
plt.scatter(y_test, lasso.predict(X_test), label='Lasso Predictions', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='black', linestyle='--')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Ridge and Lasso Regression Predictions vs Actual')
plt.legend()
plt.show()





```
